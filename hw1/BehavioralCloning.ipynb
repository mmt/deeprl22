```python
import os

import h5py
from matplotlib import pyplot
import numpy as np
import pandas as pd

train_directory = 'train'
test_directory = 'behavior_cloning_test'
models = os.listdir(test_directory)

data = {
  'names': [],
  'train_means': [],
  'train_stddev': [],
  'test_means': [],
  'test_stddev': [],
  'final_loss': [],
}
for model in models:
  f = h5py.File(os.path.join(train_directory, model), 'r')
  train_returns = np.array(f['returns'])
  f.close()
  f = h5py.File(os.path.join(test_directory, model), 'r')
  test_returns = np.array(f['returns'])
  losses = np.array(f['losses'])
  f.close()
  data['names'].append(model[:-2])
  data['train_means'].append(np.mean(train_returns))
  data['train_stddev'].append(np.var(train_returns)**0.5)
  data['final_loss'].append(losses[-1])
  data['test_means'].append(np.mean(test_returns))
  data['test_stddev'].append(np.var(test_returns)**0.5)
df = pd.DataFrame(data)
df
```
